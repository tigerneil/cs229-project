```python
import numpy as np
import pandas as pd
from importlib import reload
import util
reload(util)
data = util.FontData
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data.load()
```

```python
pca = PCA(n_components=2)

colors = {'display': 'red',
          'handwriting': 'orange',
          'monospace': 'green',
          'sans-serif': 'blue',
          'serif': 'black'}
categories = [colors[c] for c in data.get_all_category()]
marker_size = 15
```

```python
embeddings = data.get_all_embedding()
embeddings_pca = pca.fit_transform(embeddings)
```

```python
plt.scatter(embeddings_pca[:,0], embeddings_pca[:, 1], color=categories, s=marker_size)
```

```python
geometric = data.get_all_geometric()
geometric_pca = pca.fit_transform(geometric)
```

```python
plt.scatter(geometric_pca[:,0], geometric_pca[:, 1], color=categories, s=marker_size)
```

```python
semantic = data.get_all_semantic()
semantic_pca = pca.fit_transform(semantic)
```

```python
plt.scatter(semantic_pca[:,0], semantic_pca[:, 1],  color=categories, s=marker_size)
```

```python
images = data.get_all_image().reshape(1883, -1)
images_pca = pca.fit_transform(images)
```

```python
plt.scatter(images_pca[:,0], images_pca[:, 1], color=categories, s=marker_size)
```
