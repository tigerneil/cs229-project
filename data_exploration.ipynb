```python
import numpy as np
import pandas as pd
from importlib import reload
import util
reload(util)
data = util.FontData
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data.load()
```

```python
pca = PCA(n_components=2)

colors = {'display': 'red',
          'handwriting': 'orange',
          'monospace': 'green',
          'sans-serif': 'blue',
          'serif': 'black'}
categories = [colors[c] for c in data.get_all_category('train')]
marker_size = 12
```

```python
embeddings = data.get_all_embedding('train+val')
embeddings_pca = pca.fit_transform(embeddings)
```

```python
plt.scatter(embeddings_pca[:,0], embeddings_pca[:, 1], color=categories, s=marker_size)
plt.savefig("figures/data_exploration.png", dpi=200)
plt.show()
```

```python
typographic = data.get_all_typographic('train+val')
typographic_pca = pca.fit_transform(typographic)
```

```python
plt.scatter(typographic_pca[:,0], typographic_pca[:, 1], color=categories, s=marker_size)
```

```python
semantic = data.get_all_semantic('train+val')
semantic_pca = pca.fit_transform(semantic)
```

```python
plt.scatter(semantic_pca[:,0], semantic_pca[:, 1],  color=categories, s=marker_size)
```

```python
images = data.get_all_image('train+val')
images = images.reshape(images.shape[0], -1)
images_pca = pca.fit_transform(images)
```

```python
plt.scatter(images_pca[:,0], images_pca[:, 1], color=categories, s=marker_size)
```
